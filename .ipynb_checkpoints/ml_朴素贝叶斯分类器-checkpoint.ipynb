```python
# 朴树贝叶斯分类器是与线性模型非常相似的一种分类器，它的训练速度往往更快，
# 适用于维度非常高的数据集，但模型的泛化能力一般比线性分类器稍差
# sklearn中实现了三种朴树贝叶斯分类器：
# GaussianNB(高斯贝叶斯)可应用于任意连续数据
# BernoulliNB(伯努利贝叶斯)假定输入数据为二分类
# MultinomialNB(多项式贝叶斯)将输入数据为计数数据(即每个特征代表某个对象的整体计数)
# 伯努利和多项式贝叶斯主要用于文本数据分类
```

```python
# GaussianNB(高斯朴素贝叶斯),这个分类器假设每个标签的数据都服从高斯分布(正态分布)
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
X, y = make_blobs(100,2,centers=2,random_state=2,cluster_std=1.5)
plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='RdBu')
```

```python
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(X, y)

rng = np.random.RandomState(0)
Xnew = [-6,1]+[14,18]*rng.rand(50,2)
ynew = model.predict(Xnew)

plt.scatter(X[:,0], X[:,1], c=y, s=50, cmap='RdBu')
plt.scatter(Xnew[:,0], Xnew[:,1], c=ynew, marker='*', s=50, cmap='RdBu')
```
