```python
# 预处理和缩放虽然常与监督学习算法一起使用，但缩放方法并没有用到“监督”，它是无监督的
```

```python
%matplotlib inline
import mglearn

# 对数据缩放和预处理的各种方法展示
mglearn.plots.plot_scaling()
```

```python
# StandardScaler确保每个特征的平均值为0、方差为1，使所有特征都位于同一量级，
# 但这种缩放不能保证特征任何特定的最大值和最小值

# RobustScaler确保每个特征的统计属性都位于同一范围，但使用的是中位数和四分位数，
# 而不是平均值和方差，这样会忽略与其它点有很大不同的数据点(这些点也叫异常值)

# MinMaxScaler移动数据使所有特征刚好位于0到1之间

# Normalizer对每个数据点进行缩放，使得特征向量的欧式长度等于1，话句话说它将一个数据点
# 投射到半径为1的圆上(对于高维数据是球面)，这意味着每个数据点的缩放比例都不相同，
# 如果只有数据的方向(或角度)是重要的，而特征向量的长度无关紧要，常使用Normalizer归一化
```

```python
# 核SVM应用cancer数据集，并使用MinMaxScaler预处理数据
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target,
                                                   random_state=1)
print(X_train.shape)
print(X_test.shape)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
scaler.fit(X_train)
```

```python
# 调用transform()方法，返回缩放之后的数据
X_train_scaled = scaler.transform(X_train)
print("transformed shape:{}".format(X_train_scaled.shape))
print("per-feature minimum before scaling:\n{}".format(X_train.min(axis=0)))
print("per-feature maximum before scaling:\n{}".format(X_train.max(axis=0)))
print("per-feature minimum after scaling:\n{}".format(X_train_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n{}".format(X_train_scaled.max(axis=0)))
```

```python
X_test_scaled = scaler.transform(X_test)
print("per-feature minimum after scaling:\n{}".format(X_test_scaled.min(axis=0)))
print("per-feature maximum after scaling:\n{}".format(X_test_scaled.max(axis=0)))
```

```python
# 测试集缩放之后最大、最小值并不都在0~1之间，因为MinMaxScaler总是对训练集和测试集
# 进行完全相同的变换，就是说transform方法总是减去训练集的最小值，然后除以训练集的范围，
# 而这两个值可能与测试集的最小值和范围并不同
```

```python
# 为了让监督模型能够在测试集上运行，应该对训练集和测试集应用完全相同的变换
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

X, _ = make_blobs(n_samples=50, centers=5, random_state=4, cluster_std=2)
X_train, X_test = train_test_split(X, random_state=5, test_size=.1)
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
axes[0].scatter(X_train[:,0], X_train[:,1], c=mglearn.cm2(0), 
                   label="Training set", s=60)
axes[0].scatter(X_test[:,0], X_test[:,1], marker='^', c=mglearn.cm2(0), 
                   label="Training set", s=60)
axes[0].legend(loc='upper left')
axes[0].set_title("Original Data")

scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
axes[1].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=mglearn.cm2(0), 
                   label="Training set", s=60)
axes[1].scatter(X_test_scaled[:,0], X_test_scaled[:,1], marker='^', 
                c=mglearn.cm2(0), label="Training set", s=60)
axes[1].set_title("Scaled Data")

# 错误做法示例：单独对测试集进行缩放
test_scaler = MinMaxScaler()
test_scaler.fit(X_test)
X_test_scaled_badly = test_scaler.transform(X_test)
axes[2].scatter(X_train_scaled[:,0], X_train_scaled[:,1], c=mglearn.cm2(0), 
                   label="Training set", s=60)
axes[2].scatter(X_test_scaled_badly[:,0], X_test_scaled_badly[:,1], marker='^', 
                    c=mglearn.cm2(0), label="Training set", s=60)
axes[2].set_title("Improperly Scaled Data")

for ax in axes:
    ax.set_xlabel("Feature 0")
    ax.set_xlabel("Feature 1")
```

```python
# 主成分分析(PCA) 是一种旋转数据的方法，旋转后的特征在统计上不相关。
# 在做完这种旋转后，通常是根据新特征对解释数据的重要性老选择它的一个子集
```

```python
# PCA对模拟二维数据集的作用展示
mglearn.plots.plot_pca_illustration()
```

```python
# 第一张图是原始数据点，算法找到方差最大的方向，将其标记为“成分1”,
# 这是数据中包含最多信息的方向(或向量)，然后算法找到与“成分1”方向正交(成直角),
#　且包含最多信息的方向，将其标记为“成分2”，虽然这两个成分都画成箭头，
# 但其头尾位置并不重要，也可以画成反向。
# 利用这一过程找到的方向叫主成分，一般来说主成分的个数与原始特征相同
# 第二张将第一张的原始数据减去平均值，使得变换后的数据以零点为中心，
# 然后将其旋转，使第一主成分与x轴平行且第二主成分与y轴平行
# 第三张图对第二张图的数据仅保留主第一成分，将数据从二维降为一维，但是并没有保留原始特征
# 第四张图对第三张图的数据进行反向旋转并将平均值重新加到数据中，
# 这些数据点位于原始特征空间中，但是仅保留了第一主成分中包含的信息
#　这种变换有时用于除去数据中的噪声影响，或者将主成分中保留的信息可视化
```

```python
# 对cancer数据按每个特征分别计算两个类别的直方图
import numpy as np
fig, axes = plt.subplots(15, 2, figsize=(10, 20))
malignant = cancer.data[cancer.target==0]
benign = cancer.data[cancer.target==1]
ax = axes.ravel()
for i in range(30):
    _, bins = np.histogram(cancer.data[:,i],bins=50)
    ax[i].hist(malignant[:,i], bins=bins, color=mglearn.cm3(0), alpha=.5)
    ax[i].hist(benign[:, i], bins=bins, color=mglearn.cm3(2), alpha=.5)
    ax[i].set_title(cancer.feature_names[i])
    ax[i].set_yticks(())
ax[0].set_xlabel("Feature magnitude")
ax[0].set_ylabel("Frequency")
ax[0].legend(["malignant", "benign"], loc="best")
fig.tight_layout()
```

```python
# 使用PAC获取主成分
cancer = load_breast_cancer()
# 先缩放数据，使每个特征的方差为1
scaler = StandardScaler()
scaler.fit(cancer.data)
X_scaled = scaler.transform(cancer.data)
```

```python
# PAC默认情况下仅旋转(并移动)数据，但保留所有的主成分
# n_components参数可设置要保留的主成分个数
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca.fit(X_scaled)
X_pca = pca.transform(X_scaled)
print("Original shape:{}".format(X_scaled.shape))
print("Reduced shape：{}".format(X_pca.shape))
```

```python
plt.figure(figsize=(8, 8))
mglearn.discrete_scatter(X_pca[:, 0], X_pca[:, 1], cancer.target)
plt.legend(cancer.target_names, loc="best")
plt.gca().set_aspect("equal")
plt.xlabel("First principal component")
plt.ylabel("Second principal component")
```

```python
# 主成分被保存在PCA对象的components_属性中
print("PCA component shape:{}".format(pca.components_.shape))
print("PCA components:\n{}".format(pca.components_))
```

```python
plt.matshow(pca.components_, cmap='viridis')
plt.yticks([0, 1], ["First component", "Second component"])
plt.colorbar()
plt.xticks(range(len(cancer.feature_names)), cancer.feature_names,
              rotation=60, ha="left")
plt.xlabel("Feature")
plt.ylabel("Principal components")
```

```python
# 特征提取 就是可以找到一种数据表示，比给定的原始表示更适合于分析。
# 它的一个很好的应用实例就是图像，图像由像素构成，通常存储为红绿蓝(RGB)

#特征脸数据 
import matplotlib.pyplot as plt
from sklearn.datasets import fetch_lfw_people

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape
fix, axes = plt.subplots(2, 5, figsize=(15, 8), 
                           subplot_kw={'xticks':(), 'yticks':()})
for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
```

```python
# 一共有3023张图片，每张大小为87×65像素，分别属于62个不同的人
print("people.images.shape:{}".format(people.images.shape))
print("Number of classes:{}".format(len(people.target_names)))
```

```python
# 计算每个目标出现的次数
import numpy as np

counts = np.bincount(people.target)
for i, (count, name) in enumerate(zip(counts, people.target_names)):
    print("{0:25} {1:3}".format(name, count), end='  ')
    if (i + 1) % 3 == 0:
        print()
```

```python
# 为了降低数据偏斜，我们对每个人最多只取50张图像，以防止特征提取被占比高的元素影响
mask = np.zeros(people.target.shape, dtype=np.bool)
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1
X_people = people.data[mask]
y_people = people.target[mask]
# 将灰度值从0~255缩放到0~1之间，以得到更好的数据稳定性
X_people = X_people / 255.
```

```python
# K近邻用于人脸识别
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_people, y_people,
                                    stratify=y_people, random_state=0)
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train, y_train)
print("Test set score of 1-nn:{:.2f}".format(knn.score(X_test, y_test)))
```

```python
# 度量人脸的相似度，计算原始像素空间中的距离(比较每个像素的灰度值与另一张图像
# 对应位置的像素灰度值)是一种相当糟糕的方法，图像移动一个像素就得到完全不同的表示
```

```python
# 使用主成分方向的距离可以提高精度，利用PCA的白化将主成分缩放到相同的尺度，
# 变换后的结果与StandardScaler相同
# 启用白化的PCA进行数据转换示例
mglearn.plots.plot_pca_whitening()
```

```python
# 对训练数据进行拟合PCA对象，并提取前100个主成分
from sklearn.decomposition import PCA

pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)
print("X_train_pca.shape:{}".format(X_train_pca.shape))
```

```python
knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train_pca, y_train)
print("Test set sccuracy:{:.2f}".format(knn.score(X_test_pca, y_test)))
```

```python
# 非负矩阵分解(NMF) 目的在于提取有用的特征，工作原理与PCA类似，也可用于降维。
# PCA中想要的是正交分量，并且能够解释尽可能多的数据方差
# NMF中分量和系数君威非负，也就是说都大于活等于0，因此这种方法只能应用于非负数据
# 与PCA相比，NMF得到的分量更容易解释，因为负的分量和系数可能会导致难以解释的抵消效应
```

```python
# NMF在二维数据上的结果
mglearn.plots.plot_nmf_illustration()
```

```python
# 因为数据都是正的，所以数据相对于原点(0,0)的位置对NMF很重要
# 对于两个分量的NMF，如果有足够多的分量能够完美的重建数据(分量个数与特征个数相同),那么
# 算法会选择指向数据极值的方向
#　如果仅使用一个分量，NMF会创建一个指向平均值的分量，与PCA不同，减少分量个数
# 不仅会删除一些方向，而且会创建一组完全不同的分量
# NMF的分量也没有任何特定的方法排序，所有分量的地位平等
```

```python
# 分别利用10、50、100、500个分量重建不同的人脸图像
mglearn.plots.plot_nmf_faces(X_train, X_test, image_shape)
```

```python
# NMF反向变换的数据质量比PCA稍差一些，因为PCA找到的是重建的最佳方向，
# 而NMF通常并不用于对数据进行重建或编码
```

```python
# 流形学习算法  虽然PCA通常是用于变换数据的首选方法，是你能够用散点图将其可视化，
# 但这一方法的性质(先旋转然后减少方向)限制了其有效性，
# 有一类用于可视化的算法叫作流形学习算法， 它允许进行更复杂的映射，通常也可以更好可视化
#其中特别有用的一个就是t-SNE算法
```

```python
# 流形学习算法主要用于可视化，因此很好用来生成两个以上的新特征。其中一些算法(包括t-SNE)
# 计算训练数据的一种新表示，但不允许变化新数据，这意味着这些算法不能用于测试集，只能变换
# 用于训练的数据，流形学习对探索性数据分析是很有用的，但如果最终目标是监督学习则很少使用
```

```python
# t-SNE的背后思想是找到数据的一个二维表示，尽可能地保持数据点之间的距离。首先给出
# 每个数据点的随机二维表示，然后尝试让在原始特征空间中距离较近的点更加靠近，距离较远的点
# 更加远离，重点关注距离较近的点，而不是保持距离较远的点之间的距离，
# 试图保存那些表示哪些点比较靠近的信息
```

```python
# 手写数据集(每个数据点都是0~9之间首先数字的一张8×8灰度图像)应用于t-SNE
from sklearn.datasets import load_digits

digits = load_digits()

fig, axes = plt.subplots(2, 5, figsize=(10, 5), 
                             subplot_kw={'xticks':(), 'yticks':()})
for ax, img in zip(axes.ravel(), digits.images):
    ax.imshow(img)
```

```python
pca = PCA(n_components=2)
pca.fit(digits.data)
digits_pca = pca.transform(digits.data)
colors = ["#476A2A", "#7851B8", "#BD3430", "#4A2D4E", "#875525",
          "#A83683", "#4E655E", "#853541", "#3A3120", "#535D8E"]
plt.figure(figsize=(10, 10))
plt.xlim(digits_pca[:, 0].min(), digits_pca[:, 0].max())
plt.ylim(digits_pca[:, 1].min(), digits_pca[:, 1].max())
for i in range(len(digits.data)):
    plt.text(digits_pca[i, 0], digits_pca[i, 1], str(digits.target[i]),
        color=colors[digits.target[i]], fontdict={'weight':'bold', 'size':9})
plt.xlabel("First principal components")
plt.ylabel("Second principal components")
```

```python
from sklearn.manifold import TSNE

tsne = TSNE(random_state=42)
# TSNE不支持变换新数据，没有transform方法，使用fit_transform方法构建模型并返回数据
digits_tsne = tsne.fit_transform(digits.data)
plt.figure(figsize=(10, 10))
plt.xlim(digits_tsne[:, 0].min(), digits_tsne[:, 0].max())
plt.ylim(digits_tsne[:, 1].min(), digits_tsne[:, 1].max())
for i in range(len(digits.data)):
    plt.text(digits_tsne[i, 0], digits_tsne[i, 1], str(digits.target[i]),
        color=colors[digits.target[i]], fontdict={'weight':'bold', 'size':9})
plt.xlabel("t-SNE feature 0")
plt.ylabel("t-SNE feature 1")
```

```python
# PCA利用前两个分量可以将数字0、6、4较好的分开，但仍有重叠，其他数字都大量重叠在一起
# t-SNE的结果非常棒，所有的分类都被明确分开，数字1和9被分成了几块，
# 但大多数类别都形成一个密集的组
# t-SNE是无监督的,没有类别标签，但它能够找到数据的一种二维表示，
# 仅根据原始空间中数据点之间的靠近程度就能够将各个类别明确分开。
```
